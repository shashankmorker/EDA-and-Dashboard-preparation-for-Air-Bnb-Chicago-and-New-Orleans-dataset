```python
def remove_outliers(New_data, column_name):
    """
    Remove outliers from a specific column in a DataFrame using the IQR method.

    Parameters:
    - df: pandas DataFrame
    - column_name: Name of the column for outlier removal

    Returns:
    - DataFrame with outliers removed
    """

    # Calculate the first and third quartiles
    Q1 = New_data[column_name].quantile(0.25)
    Q3 = New_data[column_name].quantile(0.75)

    # Calculate the IQR (Interquartile Range)
    IQR = Q3 - Q1

    # Define the lower and upper bounds for outlier removal
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers
    New_data_no_outliers = New_data[(New_data[column_name] >= lower_bound) & (New_data[column_name] <= upper_bound)]

    return New_data_no_outliers

New_data = remove_outliers(New_data, 'price')
```
